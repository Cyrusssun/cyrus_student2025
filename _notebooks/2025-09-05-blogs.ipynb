# VS Code Setup Summary (macOS)

## 1. Installation & Configuration
- Downloaded VS Code from [code.visualstudio.com](https://code.visualstudio.com/).  
- Installed via ***.dmg*** and dragged to Applications folder.  
- Added VS Code to PATH using the command:  
  ```bash
  export PATH="$PATH:/Applications/Visual Studio Code.app/Contents/Resources/app/bin"
````

* Enabled Finder context menu options for quick folder access.

## 2. Extension Setup

* Installed **Python** extension for coding support.
* Installed **Jupyter** for notebook integration.
* Installed **GitLens** to enhance version control features.

## 3. Environment Verification

Checked Python and Git installations:

```bash
python3 --version
git --version
git config --global user.name "Your Name"
git config --global user.email "your.email@example.com"
```

## 4. Virtual Environment

Created an isolated Python environment:

```bash
python3 -m venv venv
source venv/bin/activate
```

## 5. Repository Management

* Organized personal and team repositories.
* Copied the **pages** repository to local directory.
* Experimented with themes and customizations.

## 6. Jupyter Integration

Configured notebook workflow:

```bash
jupyter notebook
git add notebook.ipynb
git commit -m "Update notebook"
git push
```

## 7. Collaboration Setup

* Set up shared repositories.
* Created personal forks to contribute to team projects.

## 8. Final Testing

* Verified all tools function correctly.
* Confirmed Python code runs in the virtual environment.
* Ensured Jupyter notebook outputs display properly.
